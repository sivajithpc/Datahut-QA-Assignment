```python
# function to fix the email format
import re
def fix_email_format(email):
    if pd.isna(email):  # Keep NaN values as is
        return None

    email = email.strip().lower()  # Trim spaces & lowercase
    
    # Regex pattern to capture missing "@" before common email providers
    pattern = r'([a-zA-Z0-9._%+-]+)(yahoo\.com|gmail\.com|hotmail\.com|outlook\.com)$'
    fixed_email = re.sub(pattern, r'\1@\2', email)  # Insert '@' before domain

    return fixed_email if '@' in fixed_email else None  # Ensure valid format

# Apply function to fix emails
data['Email'] = data['Email'].apply(fix_email_format)
def is_valid_email(email):
    pattern = r'^[a-zA-Z0-9._%+-]+@[gmail,hotmail,yahoo]+\.com'
    return bool(re.match(pattern, str(email)))

data['Email'] = data['Email'].apply(lambda x: x if is_valid_email(x) else None)
```

```python
# function to clean the names
def clean_name(name):
    if pd.isna(name):  # Keep NaN values as is
        return None
    
    name = name.strip()  # Trim spaces
    name = re.sub(r'\b(Mr|Dr|Mrs|Ms|Miss|Prof)\.? ?', '', name, flags=re.IGNORECASE)  # Remove titles
    name = re.sub(r'[^a-zA-Z\s]', '', name)  # Remove special characters
    name = re.sub(r'([a-zA-Z]+)\1$', r'\1', name)  # Remove repeated last words (e.g., Huntercold -> Hunter)
    name_parts = name.split()  # Split into words

    if len(name_parts) >= 2:
        name = f"{name_parts[0]} {name_parts[1]}"  # Keep only first and last name
    else:
        name = name_parts[0] if name_parts else ''  # Handle single-word names
    
    return name.strip()  # Return cleaned name

# Apply function to clean names
data['Name'] = data['Name'].apply(clean_name)
```

```python
# checking the combination of ID and Name to fill null values with valid names if both entries have same ID.
data["Name"] = data.groupby("ID")["Name"].transform(lambda x: x.fillna(method="ffill").fillna(method="bfill"))

data["Name"] = data["Name"].fillna("Unknown")
def get_shortest_name(group):
    valid_names = group.dropna().unique()  # Get unique non-NaN values
    if len(valid_names) == 0:
        return "Unknown"  # If no valid names exist, return default
    return min(valid_names, key=len)  # Select shortest name safely

# Step 4: Compute shortest name per ID group
shortest_names = data.groupby("ID")["Name"].apply(get_shortest_name)

# Step 5: Assign the shortest name back to the dataset
data["Name"] = data["ID"].map(shortest_names)
```
